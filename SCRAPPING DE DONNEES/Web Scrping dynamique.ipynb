```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome('C:\\Users\\chihi\\Downloads\\chromedriver_win32(4)\\chromedriver.exe')
driver.get('https://www.youtube.com/channel/UCiDy_8ig36GuMQ1FCYIv_6w/videos')
```

```python
titre= driver.find_elements_by_xpath('//*[@id="video-title"]')
vue= driver.find_elements_by_xpath('//*[@id="metadata-line"]/span[1]')
date= driver.find_elements_by_xpath('//*[@id="metadata-line"]/span[2]')

videos=[]
for i in range(len(titre)):
    video_info={}
    video_info['titre']= titre[i].text
    video_info['vue']= vue[i].text
    video_info['date']= date[i].text
    videos.append(video_info)
videos

```

```python
import pandas as pd
```

```python
df= pd.DataFrame(videos, columns=['titre', 'vue', 'date'])
```

```python
df
```

```python
df.to_csv('WebscrapingDynamique.csv')
```
