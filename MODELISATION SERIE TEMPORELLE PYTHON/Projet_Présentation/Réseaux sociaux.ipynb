# **Evolution sur différentes régions de France**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from google.colab import files
import io
uploded=files.upload()
```

```python
df_IDF=pd.read_csv(io.BytesIO(uploded["IDF.csv" ]), skiprows=1)
df_IDF.columns=['jour','instagram IDF','Facebook IDF','Snapchat IDF']
df_IDF.set_index('jour', inplace=True)
df_IDF.head()
```

```python
from google.colab import files
import io
uploded=files.upload()
```

```python
df_PACA=pd.read_csv(io.BytesIO(uploded["PACA.csv" ]), skiprows=1)
df_PACA.columns=['jour','instagram PACA','Facebook PACA','Snapchat PACA']
df_PACA.set_index('jour', inplace=True)
df_PACA.head()
```

```python
from google.colab import files
import io
uploded=files.upload()
```

```python
df_RA=pd.read_csv(io.BytesIO(uploded["RA.csv" ]), skiprows=1)
df_RA.columns=['jour','instagram RA','Facebook RA','Snapchat RA']
df_RA.set_index('jour', inplace=True)
df_RA.head()
```

```python
df_Regions = pd.concat([df_IDF, df_PACA, df_RA], axis=1)
df_Regions.head()
```

```python
df_Regions.info()
```

```python
df_Regions.plot(figsize=(20,10))
```

```python
Insta_Reg = df_Regions[['instagram IDF', 'instagram PACA', 'instagram RA']].plot(figsize=(20,10))
Insta_Reg
```

```python
Fb_Reg = df_Regions[['Facebook IDF', 'Facebook PACA', 'Facebook RA']].plot(figsize=(20,10))
Fb_Reg
```

```python
Snap_Reg = df_Regions[['Snapchat IDF', 'Snapchat PACA', 'Snapchat RA']].plot(figsize=(20,10))
Snap_Reg
```
